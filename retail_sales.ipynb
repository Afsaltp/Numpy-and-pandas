```python
import pandas as pd
```

```python
df=pd.read_csv('retail_sales_dataset.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
print(df.isnull().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df['UnitPrice'] = df['UnitPrice'].fillna(df['UnitPrice'].median())
```

```python
df.info()
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Product Category counts

```python
plt.figure(figsize=(8,5))
sns.countplot(x="ProductCategory", data=df)
plt.title("Product Category Distribution")
plt.xticks(rotation=45)
plt.show()
```

# 2. Histogram – Age distribution

```python
plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

# 3. Scatter Plot – Unit Price vs Quantity

```python
plt.figure(figsize=(8,5))
plt.scatter(df['UnitPrice'], df['Quantity'])
plt.title("Unit Price vs Quantity")
plt.xlabel("Unit Price")
plt.ylabel("Quantity")
plt.show()
```

# 4. Box Plot – Price per Category

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="ProductCategory", y="UnitPrice", data=df)
plt.title("Unit Price per Category")
plt.xticks(rotation=45)
plt.show()
```

**#CONCLUSION**

**The dataset shows that most purchases come from Electronics and Clothing, making them the most in-demand categories. The majority of customers fall between the ages of 25–45, indicating that the store mainly attracts working-age adults.**
