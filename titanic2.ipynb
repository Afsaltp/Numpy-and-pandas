```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
df.columns
```

```python
df.info()
```

```python
df['fam_count']=df['sibsp']+df['parch']
```

```python
df.head()
```

```python
df=df.drop(columns=['sibsp','parch'])
df
```

```python
df['age']=df['age'].round(1)
```

```python
df['fare']=df['fare'].round(1)
```

```python
df['sex']=df['sex'].str.title()# str is an attribute which will help us to apply str built in methods it helps fetch sinngle single value
```

```python
df=df.rename(columns={'sex':'gender'})# rename use to change the column
df
```

```python
df['embarked']=df['embarked'].replace({'S':'Southampton','C':'Cherbourg','Q':'Queestown'})# to replace values
df
```

```python
df[['lname','temp']]=df['name'].str.split(',',expand=True)
df[['title','fname']]=df['temp'].str.split('.',expand=True,n=1)
df=df.drop(columns=['name','temp'])
for col in ['lname','fname','title']:
    df[col].str.strip()
```

```python
for col in ['lname','fname','title']:
    df[col]=df[col].str.strip()
df
```

```python
ticket_cnt=df['ticket'].value_counts()# it acts like aggreagate function that ticket  is index values are the vales for index
# return type is series
```

```python
ticket_cnt=ticket_cnt.reset_index() # data frame
```

```python
df=pd.merge(df,ticket_cnt,'inner',on='ticket')#pd.merge(lefttab,right tab,how='inner',on='colname')
df
```

```python
df=df.rename(columns={'count':'passenger_cnt'})
df.columns
```

```python
l=['ticket','title','fname','lname','gender', 'age','embarked','fam_count','passenger_cnt','pclass', 'fare', 'survived']
df=df[l]
```
