```python
top10 = df.nlargest(10, 'CreditScore').sort_values('CreditScore')
top10
```

```python
#1.Top 10 Highest CreditScore
plt.figure(figsize=(8,5))
plt.barh(top10['Surname'], top10['CreditScore'])
plt.xlabel("Credit Score")
plt.ylabel("Surname")
plt.title("Top 10 Highest Credit Scores")
plt.tight_layout()
plt.show()
```

1.All top 10 customers have the same Credit Score of 850, indicating they all belong to the highest credit-quality group in the dataset.

2.These customers come from different surnames, showing no single family or group dominates the highest score range.

3.Since credit scores are identical, the ranking is based only on data order, not score differences  meaning credit score alone cannot differentiate these customers.

```python
df['Exited']=df['Exited'].replace({1:"Account_Closed",0:'Account_Open'})
df['Exited']
```

```python
#Age Category vs Churn
plt.figure(figsize=(6,5))
sns.countplot(data=df, x="Age_Category", hue="Exited")

plt.title("Churn Count Across Age Categories")
plt.xlabel("Age Category")
plt.ylabel("Number of Customers")

plt.show()

```

1.Adult customers have the highest count, and most of them keep their accounts open.

2.Middle Age customers show the highest number of account closures, indicating more churn in this group.

3.Very few customers in the Young and Senior groups close their accounts, showing low churn among them.

```python
# Credit Score vs Churn
plt.figure(figsize=(6,5))
sns.boxplot(data=df, x="Exited", y="CreditScore")

plt.title("Credit Score Distribution for Churned vs Retained Customers")
plt.xlabel("Exited (0 = No, 1 = Yes)")
plt.ylabel("Credit Score")

plt.show()
```

1.Both Account Open and Account Closed customers have a very similar credit score distribution, meaning credit score does NOT strongly impact churn.

2.The median credit score of both groups is almost the same, showing no major difference in credit quality.

3.A few low credit score outliers appear in the Account Closed group, but overall the trend is nearly identical.

```python
#Number of Products vs Churn
plt.figure(figsize=(6,5))
sns.countplot(data=df, x="NumOfProducts", hue="Exited")

plt.title("Churn Distribution by Number of Bank Products")
plt.xlabel("Number of Products")
plt.ylabel("Number of Customers")

plt.show()
```

1.Most customers have 1 or 2 bank products, and these groups also show the highest churn.

2.Customers with 3 or more products rarely churn, indicating multi-product users are more loyal.

3.The highest retention is seen for customers with 2 products, showing stronger engagement.

```python
#Credit Card vs Active Membe
plt.figure(figsize=(6,5))
sns.countplot(data=df, x="HasCrCard", hue="IsActiveMember")

plt.title("Active Status Comparison Between Credit Card Holders")
plt.xlabel("Credit Card Status (CreditCard / NoCreditCard)")
plt.ylabel("Number of Customers")

plt.show()
```

1.Customers with a credit card are more active compared to those without a credit card.

2.Even among non-credit-card users, active and not-active counts are close, but overall numbers are smaller.

3.Credit card holders show higher engagement, suggesting they are more involved with the bank's services.

```python
#Estimated Salary vs Churn
plt.figure(figsize=(6,5))
sns.boxplot(data=df, x="Exited", y="EstimatedSalary")

plt.title("Estimated Salary Distribution for Churned vs Retained Customers")
plt.xlabel("Exited (0 = No, 1 = Yes)")
plt.ylabel("Estimated Salary")

plt.show()
```

1.The salary distribution is almost identical between customers who closed their accounts and those who stayed.

2.Both groups have a similar median salary, showing salary does not strongly influence churn.

3.High-salary and low-salary customers appear in both groups, indicating churn occurs across all income levels.

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(), cmap="coolwarm", annot=True, fmt=".2f")

plt.title("Correlation Heatmap of Numeric Features")
plt.xlabel("Features")
plt.ylabel("Features")

plt.show()
```

# CONCLUTION:

1.Most numerical features show very weak correlations with each other, meaning the dataset is not strongly linear.

2.The strongest relationship is a negative correlation (~0.30) between Balance and Number of Products  customers with more products tend to have lower balance.

3.CreditScore, Age, Salary, and Tenure have almost no correlation with each other, indicating these features vary independently.
