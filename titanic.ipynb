```python
age_mean=df['age'].mean()
age_mean
```

```python
fare_mean=df['fare'].mean()
fare_mean
```

```python
df['age']=df['age'].fillna(age_mean)
df.info()
```

```python
df['fare']=df['fare'].fillna(fare_mean)
df.info()
```

```python
emb_mod=df['embarked'].mode()[0]
```

```python
df['embarked']=df['embarked'].fillna(emb_mod)
df.info()
```

```python
cab_mod=df['cabin'].mode()[0]
cab_mod
```

```python
df['cabin']=df['cabin'].fillna(cab_mod)
df.info()
```

```python
df=df.drop(columns=['cabin'])
```

```python
df.info()
```

```python
df.sort_values(by="ticket")
```

```python
df=df.sort_values(by="ticket").reset_index(drop=True)
df
```

```python
df=df.rename(columns={'sex' : 'gender'})
df
```

```python
df['fam_count']=df['parch']+df['sibsp']
df['fam_count'].head()
```

```python
df['individual_fare']=df['fare']/df['psg count']
df['individual_fare']
```

```python
df['psg count']=df['fam_count']+1
df['psg count']
```

```python
l=['ticket','name','gender','age','embarked','psg count','fam_count','pclass','fare','individual_fare','survived']
df=df[l]
df
```
