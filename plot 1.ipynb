```python
import pandas as pd
import numpy as np
import numpy.random as rd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'year': [2015,2015,2016,2016,2017,2017,2018,2018,2019,2019,2020,2020,2021,2021,2022],
    'product': ['A','B','A','B','A','B','A','B','A','B','A','B','A','B','B'],
    'sales': [120,480,135,460,128,490,150,520,140,530,155,550,160,580,1500],
    'marketing_spend': [12,42,15,40,13,45,18,47,14,48,17,50,19,53,90],
    'customer_rating': [3.8,4.2,2.9,3.5,4.5,3.8,2.7,4.7,4.9,3.2,3.0,4.8,2.8,4.1,2.5]
}

df = pd.DataFrame(data)

df
```

```python
plt.figure(figsize=(3,3))
sns.lineplot(df,x='year',y='sales')
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.lineplot(df,x='year',y='sales',label='sales') #blue
sns.lineplot(df,x='year',y='marketing_spend',label='marketing')
plt.show()
```

```python
fig,axes=plt.subplots(ncols=3,figsize=(9,3))
sns.lineplot(df,x='year',y='sales',label='sales',ax=axes[0]) 
sns.lineplot(df,x='year',y='marketing_spend',label='marketing',ax=axes[1])
sns.lineplot(df,x='year',y='customer_rating',label='rating',ax=axes[2])
```

```python
fig,axes=plt.subplots(ncols=3,figsize=(9,3))
sns.lineplot(df,x='year',y='sales',hue='product',ax=axes[0]) 
sns.lineplot(df,x='year',y='marketing_spend',hue='product',ax=axes[1])
sns.lineplot(df,x='year',y='customer_rating',hue='product',ax=axes[2])
plt.tight_layout()
plt.show()
```
