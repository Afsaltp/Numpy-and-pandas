```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("music_streaming_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Fill missing values with mode (most common value)
df['artist_name'].fillna(df['artist_name'].mode()[0], inplace=True)
df['genre'].fillna(df['genre'].mode()[0], inplace=True)
df['user_rating'].fillna(df['user_rating'].mean(), inplace=True)
df['country'].fillna(df['country'].mode()[0], inplace=True)
```

```python
# drop missing values
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['genre'] = df['genre'].str.title()    
df['platform'] = df['platform'].str.title()   
df['playlist_inclusion'] = df['playlist_inclusion'].str.title()
```

```python
print(df.describe())
```

```python
# Top 10 Most Streamed Songs
top_songs = df.sort_values(by='stream_count', ascending=False).head(10)
plt.figure(figsize=(6,3))
sns.barplot(x='stream_count', y='song_name', data=top_songs, palette='Blues_d')
plt.title("Top 10 Most Streamed Songs")
plt.xlabel("Stream Count")
plt.ylabel("Song Name")
plt.show()
```

Song_268 is the top first song followed by Song_191 and then Song_298

```python
# Distribution of User Ratings
plt.figure(figsize=(4,3))
sns.histplot(df['user_rating'], bins=10, color='orange')
plt.title("Distribution of User Ratings")
plt.xlabel("User Rating")
plt.ylabel("Count")
plt.show()
```

User distribution shows that most users rated average(3 rating)

```python
# Most Popular Genres
plt.figure(figsize=(6,4))
sns.countplot(x='genre', data=df, order=df['genre'].value_counts().index, palette='Set2')
plt.title("Most Popular Genres")
plt.xlabel("Genre")
plt.ylabel("Number of Songs")
plt.xticks(rotation=45)
plt.show()
```

the dataset contains more EDM songs

```python
# Average Rating by Platform
plt.figure(figsize=(6,3))
sns.barplot(x='platform', y='user_rating', data=df, ci=None, palette='coolwarm')
plt.title("Average User Rating by Platform")
plt.xlabel("Platform")
plt.ylabel("Average Rating")
plt.show()
```

Spotify and Amazon Music have the highest average ratings (around 3.0), showing strong user satisfaction

```python
# Skip Rate vs User Rating (Relationship)
plt.figure(figsize=(8,5))
sns.scatterplot(x='skip_rate', y='user_rating', data=df, color='green')
plt.title("Skip Rate vs User Rating")
plt.xlabel("Skip Rate (%)")
plt.ylabel("User Rating")
plt.show()
```

The points are spread out widely, showing no strong direct pattern between skip rate and rating.

```python
# Average Stream Count per Genre
plt.figure(figsize=(6,3))
genre_stream = df.groupby('genre')['stream_count'].mean().sort_values(ascending=False).reset_index()
sns.barplot(x='genre', y='stream_count', data=genre_stream, palette='rainbow')
plt.title("Average Stream Count per Genre")
plt.xlabel("Genre")
plt.ylabel("Average Stream Count")
plt.xticks(rotation=45)
plt.show()
# viridis
```

EDM is the most streamed genre

# CONCLUSION
The analysis shows that users mostly give average ratings, prefer modern genres like EDM, and rate Spotify and Amazon Music the highest. Skip rates are generally higher for lower-rated songs, indicating that user engagement decreases with song quality.
