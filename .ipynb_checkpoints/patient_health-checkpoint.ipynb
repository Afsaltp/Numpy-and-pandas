```python
import pandas as pd
```

```python
df=pd.read_csv("patient_health_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df["BloodPressure"]=df["BloodPressure"].fillna(df["BloodPressure"].mean())
df["CholesterolLevel"]=df["CholesterolLevel"].fillna(df["CholesterolLevel"].median())
df["HeartRate"]=df["HeartRate"].fillna(df["HeartRate"].median())
```

```python
df=df.drop_duplicates()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Age distribution

```python
plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=12)
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

Most patients are between 30 and 65 years old.

# 2. Blood Pressure vs Age

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Age"], df["BloodPressure"])
plt.title("Age vs Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.show()
```

The scatter plot shows that blood pressure tends to increase as age increases.

# 3. Cholesterol Levels by Diabetes

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="HasDiabetes", y="CholesterolLevel", data=df)
plt.title("Cholesterol Levels: Diabetic vs Non-Diabetic")
plt.xticks([0, 1], ["No Diabetes", "Diabetes"])
plt.show()
```

Diabetic patients usually have higher cholesterol levels than non-diabetic patients.

# 4. Exercise vs Heart Disease

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="ExerciseHours_PerWeek", y="HasHeartDisease", data=df)
plt.title("Exercise Hours vs Heart Disease")
plt.show()
```

Patients who exercise fewer hours per week show a slightly higher chance of having heart disease.

# 5. Correlation Heatmap

```python
df_numeric = df.select_dtypes(include=["int64", "float64"])
plt.figure(figsize=(8,5))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

Strong correlations appear between age and blood pressure, and between diabetes and cholesterol levels.

# CONCLUTION

The dataset highlights clear health patterns among patients. Most individuals fall within the 3065 age range, and older patients tend to show higher blood pressure values. Cholesterol levels are noticeably higher in diabetic patients compared to non-diabetics.
