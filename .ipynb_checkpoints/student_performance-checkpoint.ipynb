```python
import pandas as pd
```

```python
df = pd.read_csv("student_performance_dataset_final.csv")
```

```python
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Attendance"]=df["Attendance"].fillna(df["Attendance"].mean())
```

```python
df["Hours_Studied"]=df["Hours_Studied"].fillna(df["Hours_Studied"].median())
```

```python
df["FinalMarks"]=df["FinalMarks"].fillna(df["FinalMarks"].median())
```

```python
df=df.drop_duplicates()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Department Count

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Department", data=df)
plt.title("Student Count by Department")
plt.xticks(rotation=45)
plt.show()
```

The bar chart shows how students are distributed across different departments. CSE and ECE usually have the highest number of students

# 2. Age Distribution

```python
plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=10)
plt.title("Age Distribution of Students")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

The histogram shows that most students are between 18 and 21 years old.

# 3. Study Hours vs Final Marks

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Hours_Studied"], df["FinalMarks"])
plt.title("Study Hours vs Final Exam Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Final Marks")
plt.show()
```

students who study more hours generally achieve higher final marks.

# 4. Attendance vs Final Marks

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Attendance", y="FinalMarks", data=df)
plt.title("Attendance vs Final Marks")
plt.show()
```

This plot shows that students with higher attendance usually score higher in final exams.

# 5. Correlation Heatmap

```python
df_numeric = df.select_dtypes(include=["int64", "float64", "float32"])

plt.figure(figsize=(8,5))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

The heatmap reveals that both study hours and attendance are moderately correlated with final marks

# CONCLUTION

The dataset shows that most students come from CSE and ECE departments and fall within the 18â€“21 age group. Academic performance is strongly shaped by behaviors like study hours and attendance, both of which show clear positive relationships with final exam marks.
